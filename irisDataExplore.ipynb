```python
import pandas as pd
```

### 1. IrisData.csv 파일 읽기

```python
data=pd.read_csv('./IrisData.csv')
```

### 2. 데이터 살펴보기

```python
#데이터 구조 확인
print("Shape = ",data.shape)
print("Index = ",data.index)
print("Columns = ",data.columns)
```

```python
#데이터 내용 일부 확인
display(data.head())
display(data.tail())
```

```python
#통계요약 확인
data.describe()
```

```python
#누락된 데이터 확인
miss=data.isna()
print("누락된 데이터 갯수 =\n",miss.sum())
```

```python
#Iris의 Species 속성값 확인
print("데이터 타입 = \n", data['Species'].dtypes)
print("행 레이블 = \n", data['Species'].index)
print("각 그룹 = \n", data.groupby('Species').size())
```

```python
#Iris의 꽃잎 길이(Petal length)의 역순(=꽃잎길이가 작은 순)으로 확인
display(data.sort_values(by='Petal length'))
```

### 3. 데이터 추출

```python
#Species(종)이 setosa인 경우에만 추출
sp=data.groupby('Species')
setosa=sp.get_group(name='setosa')
display(sp.get_group(name='setosa'))
```

```python
#setosa에서 꽃받침길이(Sepal length)가 5이하인 데이터만 추출
sl=setosa.loc[:,'Sepal length']<=5
display(setosa[sl])
```

```python
#setosa에서 꽃받침 폭 속성, 길이 속성만 추출
setosa.loc[:,'Sepal length':'Sepal width']
```

### 4. 데이터 추가

```python
#꽃받침 길이( Sepal length)와 꽃받침 폭( Sepal width)의 비율을 새로운 속성으로 추가
s=data.loc[:,'Sepal length']/data.loc[:,'Sepal width']
data['Sepal ratio']=s
display(data)
```

```python
#꽃잎 길이(Petal length) 와 꽃잎 폭(Petal width)의 비율을 새로운 속성으로 추가
p=data.loc[:,'Petal length']/data.loc[:,'Petal width']
data['Petal ratio']=p
display(data)
```

### 5. 그룹별 분리, 샘플링, 시각화

```python
#Species을 기준으로 groupby처리
sg=data.groupby('Species')
```

```python
#Species별 평균, 표준편차 확인
display("평균 = ", sg.mean())
display("표준편차 = ", sg.std())
```

```python
#Species(종별)로 분리
spe={}
for name, species in sg:
    print("Species name : ", name)
    spe[name]=species
    display(species)
```

```python
#분리된 3개의 종에서 각각 20개씩 랜덤 추출하여 결합
s1=spe['setosa'].sample(20)
s2=spe['versicolor'].sample(20)
s3=spe['virginica'].sample(20)
s1=s1.append(s2)
s1=s1.append(s3)
display(s1)
```

```python
#3개 종에 대한 boxplot (subplot 이용)
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(20, 10))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
spe['setosa'].boxplot(ax=ax1)
spe['versicolor'].boxplot(ax=ax2)
spe['virginica'].boxplot(ax=ax3)
plt.show()
```

```python
#3개 종에 대한 histogram
fig=plt.figure(figsize=(20, 10))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
spe['setosa'].plot.hist(ax=ax1)
spe['versicolor'].plot.hist(ax=ax2)
spe['virginica'].plot.hist(ax=ax3)
plt.show()
```

```python
#3개 종에 대한 scatter plot을 결합
#Sepal width, Petal width 이용
ax=spe['setosa'].plot.scatter(x='Sepal width',y='Petal width', color='Violet', label='setosa')
ax=spe['versicolor'].plot.scatter(x='Sepal width',y='Petal width', color='Green', label='versicolor', ax=ax)
ax=spe['virginica'].plot.scatter(x='Sepal width',y='Petal width', color='Orange', label='virginica', ax=ax)
```

```python
#Sepal length, Petal length 이용
ax=spe['setosa'].plot.scatter(x='Sepal length',y='Petal length', color='Violet', label='setosa')
ax=spe['versicolor'].plot.scatter(x='Sepal length',y='Petal length', color='Green', label='versicolor', ax=ax)
ax=spe['virginica'].plot.scatter(x='Sepal length',y='Petal length', color='Orange', label='virginica', ax=ax)
```

```python
#(4)의 결합된 데이터(=s1)의 scatter matrix
pd.plotting.scatter_matrix(s1, s=60, diagonal='kde')
```

```python
#(4)의 결합된 데이터(=s1)의 평균에 대한 bar, pie plot
mean=s1.mean()
display(mean)
mean.plot.bar(title='bar plot', subplots=ax, figsize=(8, 4))
plt.show()
mean.plot.pie(title='pie plot', autopct='%.2f%%', subplots=True, figsize=(8, 8))
plt.show()
```
