```python
#단순 데이터 (make_wav)를 사용한 단순선형회귀 예 (강의자료 6,7쪽)
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

X, y = mglearn.datasets.make_wave(n_samples=60)#인위적인 데이터 집합
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)#데이터셋을 훈련용과 테스트용으로 나누기

lr = LinearRegression().fit(X_train, y_train)

w = lr.coef_#가중치
b = lr.intercept_#편향치
print("w (lr.coef_) : ", w)
print("b (lr.intercept) : ", b)

print("훈련 세트 점수 : {:.2f}".format(lr.score(X_train, y_train)))
print("데스트 세트 점수 : {:.2f}".format(lr.score(X_test, y_test)))
```

```python
line = np.linspace(-3, 3, 100).reshape(-1, 1)
plt.figure(figsize=(8, 8))
plt.plot(line, lr.predict(line))#선형회귀로 추정된 선형함수(모델)로 출력값 생성
plt.plot(X, y, 'o', c='blue')#학습 데이터
plt.ylim(-3, 3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Wave Data Plot")
plt.legend(["model", "train data"], loc="best")
plt.grid()
plt.show()
```

```python
#보스톤 주택가격(1978) 데이터 집합을 사용한 다중선형회귀 예 (강의자료 8쪽)
import sklearn.datasets as ds

boston = ds.load_boston()

print("데이터의 형태 : ", boston.data.shape)
print("특성 이름 : \n", boston.feature_names)
print("데이터 설명 : \n", boston.DESCR)

X = boston.data
y = boston.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)

print("훈련 세트 점수 : {:.2f}".format(lr.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(lr.score(X_test, y_test)))
```

```python
#데이터 집합에서 주택가격과 개별특징의 상관도 확인 (강의자료 10쪽)
import pandas as pd
df = pd.DataFrame(boston.data, columns=boston.feature_names)#데이터 프레임 작성
df['MEDV'] = boston.target#주택가격

df.hist(bins=10, figsize=(9, 7), grid=False)

corr = df.corr(method='pearson')#개별 특증들간의 상관도
display(corr)
corrMEDV = corr['MEDV']#주택가격의 상관도
display(corrMEDV)
sub = corrMEDV[corrMEDV.abs() > 0.45]
display(sub)
```

```python
#보스톤 주택가격(1978) 데이터 집합에서 주택가격과 높은 상관도를 가지는 특징들로 다중선형회귀 결과 확인 (강의자료 12쪽)
d=df.loc[:,['INDUS', 'RM', 'TAX', 'PTRATIO', 'LSTAT']]
X = d
y = df['MEDV']
print(d.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lrt = LinearRegression().fit(X_train, y_train)

w = lrt.coef_#가중치
b = lrt.intercept_#편향치
print("w (lr.coef_) : ", w)
print("b (lr.intercept) : ", b)

print("훈련 세트 점수 : {:.2f}".format(lrt.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(lrt.score(X_test, y_test)))
```

```python
#Ridge Regression 예(강의자료 15-16쪽)
from sklearn.linear_model import Ridge

X = boston.data
y = boston.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

ridge = Ridge().fit(X_train, y_train)
w = ridge.coef_
b = ridge.intercept_
print("---------------------------")
print("Ridge(alpha=1)")
print("w (ridge.coef_) : ",w)
print("b (ridge.intercept_) : )", b)

print("훈련 세트 점수 : {:.2f}".format(ridge.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(ridge.score(X_test, y_test)))

ridge10 = Ridge(alpha=10).fit(X_train, y_train)
w = ridge10.coef_
b = ridge10.intercept_
print("---------------------------")
print("Ridge(alpha=10)")
print("w (ridge10.coef_) : ",w)
print("b (ridge10.intercept_) : )", b)

print("훈련 세트 점수 : {:.2f}".format(ridge10.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(ridge10.score(X_test, y_test)))

ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
w = ridge01.coef_
b = ridge01.intercept_
print("---------------------------")
print("Ridge(alpha=0.1)")
print("w (ridge10.coef_) : ",w)
print("b (ridge10.intercept_) : )", b)

print("훈련 세트 점수 : {:.2f}".format(ridge01.score(X_train, y_train)))
print("테스트 세트 점수 : {:.2f}".format(ridge01.score(X_test, y_test)))
```

```python
#alpha 값에 따른 계수 비교
ind = boston.feature_names
dfe = pd.DataFrame({'Ridge alpha=10' : ridge10.coef_,
                   'Ridge alpha=1' : ridge.coef_,
                   'Ridge alpha=0.1' : ridge01.coef_,
                   'LinearRegression' : lr.coef_}, index=ind)
dfe.plot.bar(figsize=(10, 10))
plt.xlabel("features")
plt.ylabel("coef")
plt.ylim(-20, 10)
plt.legend(fontsize=12, loc='buttom right')
```
