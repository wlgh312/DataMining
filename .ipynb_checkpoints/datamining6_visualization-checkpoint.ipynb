## simple plot

```python
import numpy as np
import matplotlib.pyplot as plt

#plot()     
x = np.linspace(0, 3, 100) #0 ~ 3    100  
x1 = x
x2 = x**2
x3 = x**3
plt.plot(x, x1, label='linear', color='r')    #1  (x=x, y=x1), 
plt.plot(x, x2, label='quadratic', color='g') #2 (x=x, y=x2), 
plt.plot(x, x3, label='cubic', color='b')     #3 (x=x, y=x3), 
plt.xlabel('x label')  #x 
plt.ylabel('y label')  #y 
plt.title('Simple plot') #
plt.legend()  #
plt.show()  # 
```

```python
h = [160, 180, 177, 156, 187, 170]
w = [50, 100, 66, 55, 75, 70]
plt.plot(h, w, 'bo')  # plot x and y using blue circle markers
plt.show()
```

### Bar plot
  ,        


 (categorical variable) : , ,        (discrete)    

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display

# 
df1 = pd.DataFrame({'lab':['A', 'B', 'C'], 'val':[10, 30, 20]})
df1.plot.bar(x='lab', y='val', title="Bar plot",rot=0) #x :  , y:  (), title:, rot : x  
plt.show()

# 
speed = [0.1, 17.5, 40, 48, 52, 69, 88]
lifespan = [2, 8, 70, 1.5, 25, 12, 28]
index = ['snail', 'pig', 'elephant',
          'rabbit', 'giraffe', 'coyote', 'horse']
df2 = pd.DataFrame({'speed': speed,
                   'lifespan': lifespan}, index=index)
display(df2)
df2.plot.bar(rot=0)
plt.show()
# y , x   
df2.plot.bar(y=['lifespan', 'speed'], rot=45) 
plt.show()
# y 
df2.plot.bar(y=['speed'], rot=0) 
plt.show()
#subplot  
axes = df2.plot.bar(rot=0, subplots=True,  figsize=(8, 4))   #,   (8,  4)
plt.show()
```

### Pie plot

  ,        


```python
df = pd.DataFrame(3 * np.random.rand(4, 2),
                  index=['a', 'b', 'c', 'd'], columns=['x', 'y'])
display(df)

# x 
df.plot.pie(y='x', title = 'pie plot (x)')  #x 
# y 
df.plot.pie(y='y', title = 'pie plot (y)',  autopct='%.2f%%')  #y, %  

#x, y
df.plot.pie( title = 'pie subplot', subplots=True, figsize=(8, 4))  

# df2  
df2.plot.pie( title = 'pie plot (speed-lifespan)', subplots=True, figsize=(14, 8))

plt.show()
```

```python
df3 = pd.DataFrame({
     'x': [1, 2, 2.5, 3, 3.5, 4, 5],
     'y': [4, 4, 4.5, 5, 5.5, 6, 6],
 })

#histogram
df3.plot.hist()
df3.plot.hist( subplots=True)
df3.plot.hist(y='x')
df3.plot.hist(y='x', bins=3)

# histogram 
df3.hist()
df3.hist(bins=3)

plt.show()
```

### density  
   Probability Density  Function, pdf)  
  ( kernel density estimation,KDE) 

```python
df3 = pd.DataFrame({
     'x': [1, 2, 2.5, 3, 3.5, 4, 5],
     'y': [4, 4, 4.5, 5, 5.5, 6, 6],
 })

df3.hist(bins=3)

#Generate Kernel Density Estimate plot using Gaussian kernels.
df3.plot.density()
df3.plot.kde()
plt.show()
```

```python
#plt  Histogram,     

mu, sigma = 100, 15
x = np.random.normal(mu, sigma, size=1000) # 60,  20  1000 

n, bins, patches = plt.hist(x, bins=10)
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Histogram1')
plt.text(120, 200, '$ \mu=100,\ \sigma=15 $') 
plt.show()

print ("n=", n)   #
print ("bins=", bins) #


sx = pd.Series(x) #x  pandas Series 

sx.plot.hist()
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Histogram2')
plt.text(120, 200, '$ \mu=100,\ \sigma=15 $') 
plt.show()

sx.plot.kde()
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Histogram2')
plt.show()
```

```python
#  2 
df2 = pd.DataFrame([[7.1, 3.4, 0], [6.7, 4.0, 0], [5.6, 9.3, 1],
                  [5.3, 7.8, 1], [5.8, 4.6, 2],[2.5, 3.2, 2]],
                  columns=['length', 'width', 'types'])
display(df2)
df2.plot.scatter(x='length',  y='width',   c='Blue')  #c 
df2.plot.scatter(x='length',  y='width',  c='types', colormap='plasma') #type   
df2.plot.scatter(x='length',  y='width',   marker='^', s= 60 ) #  
df2.plot.scatter(x='length',  y='width', s= df2['types'] * 50) #type  

#  4 
df3 = pd.DataFrame(np.random.rand(20, 4) * 5, columns=['a', 'b', 'c', 'd'])
display(df3)
ax = df3.plot.scatter(x='a', y='b', color='Skyblue', label='Group 1');
ax = df3.plot.scatter(x='c', y='d', color='Orange', label='Group 2', ax=ax); #ax plot  

plt.show()
```

```python
#scatter plot ,   
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
display(data)

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

### Scatter Matrix
   scatter plot matrix

```python
df4 = pd.DataFrame(np.random.rand(20, 4) * 5, columns=['a', 'b', 'c', 'd'])

df4['class'] =np.round(np.random.rand(20,1) * 5)  #class  
df4['class']= df4['class'].astype("category")     # 

display(df4)

pd.plotting.scatter_matrix(df4, s=60, diagonal='kde', c= df4['class'], cmap='spring') #class spring 
pd.plotting.scatter_matrix(df4, s=60, diagonal='hist', c= df4['class'], cmap= 'viridis')
plt.show()
```
