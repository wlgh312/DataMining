### Kaggle이란?
예측 모델 및 분석 대회 플랫폼이다. 기업 및 단체에서 데이터와 해결과제를 kaggle을 통해 등록하면, 여러 사람들이 이를 해결하는 모델을 개발하고 경쟁한다.  
기업이 Kaggle에게 빅데이터를 제공해주고 Kaggle은 이를 온라인에 공개하여 세계 각지의 수많은 데이터 과학자들로 하여금 이 문제를 팀이나 개인으로 해결할 수 있도록 해준다. 즉, 캐글은 세계 최대 데이터 과학자 커뮤니티인 셈이다.

```python
#Boston House Prices dataset
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('./HousingData.csv')
```

* CRIM : 마을에서 1인당 범죄율  
* ZN : 25,000평방 피트가 넘는 부지에 대한 주거용 토지의 비율  
* INDUS : 마을 당 비소매 사업지역의 토지 비율  
* CHAS : 찰스강 더미 변수 (강의 경계에 위치 = 1, 아니면 = 0)  
* NOX : 10ppm당 농축 일산화질소  
* RM : 주택 1가구 당 방의 평균 개수  
* AGE : 1940년 이전에 건축된 소유자 주택 비율  
* DIS : 5개의 보스턴 고용센터까지의 접근성 지수  
* RAD : 방사형 도로까지의 접근성 지수  
* TAX : 10,000달러 당 재산세율  
* PTRATIO : 타운별 학생/교사 비율  
* B : 마을 당 흑인의 비율
* LSTAT : 모집단위 하위계층의 비율
* MEDV : 본인 소유의 주택가격(중앙값)  

```python
#데이터의 구조 확인
print("Types = ", data.dtypes)
print("Shape = ", data.shape)
print("Index = ", data.index)
print("Columns = ", data.columns)
```

```python
#기초통계요약 확인
data.describe()
```

```python
#모든 특징들에 대한 상관계수 확인
data.corr(method='pearson')
```

```python
#주택가격(medv)과 다른 특징들과의 상관계수 확인
data.corrwith(data['MEDV'])
```

```python
#RM, PTRATIO, LSTAT, MEDV 4개 속성만 사용
#4개 속성에 대한 histogram, boxplot 확인
fdata=data[['RM','PTRATIO','LSTAT','MEDV']]
fdata.plot.hist()
plt.show()
fdata.boxplot()
plt.show()
```

```python
#4개 속성에 대한 scatter plot (산점도 행렬 : Scatterplot Matrices)
#다른 특징들과 집의 가격을 비교하는게 맞다고 생각해 y를 'MEDV'로 지정했습니다.
fdata.plot.scatter(x='RM', y='MEDV')
fdata.plot.scatter(x='PTRATIO', y='MEDV', color='Green')
fdata.plot.scatter(x='LSTAT', y='MEDV', color = 'Red')
```
