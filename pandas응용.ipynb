```python
import numpy as np
import pandas as pd
from IPython.display import display
```

### 1. ageGroup(1-9),sex(1,2),height,weight,BMI,bodyFat으로 구성된 파일(physicalFitness.csv)를 읽고 데이터프레임 만들기

```python
data=pd.read_csv('./physicalFitness.csv')
pd.DataFrame(data)
```

### 2.  (1)의 데이터 셋에 대한 데이터 타입들, 요약통계, 상,하위 5개 데이터 확인, 행 레이블, 열레이블 , 크기 확인

```python
print("Data Type =\n{}\n".format(data.dtypes))
print("통계요약 =\n{}\n".format(data.describe()))
print("상위 5개 데이터 =\n",data.head(5))
print("\n하위 5개 데이터 = \n",data.tail(5))
print("\nRow Labels = ",data.index)
print("\nColumns Lables = ",data.columns)
print("\nData Size = ",data.size)
```

### 3. 키가 큰 순으로 정렬

```python
data.sort_values(by='height', ascending=False)
```

### 4.  BMI 30이상인 비만자만 ageGroup,height,weight,BMI,bodyFat을 추출하여 파일(obese.csv)로 저장

```python
bmi=data['BMI']>=30
obese=data[bmi].loc[:,['ageGroup','height', 'weight', 'BMI', 'bodyFat']]
obese.to_csv('./obese.csv')
```

### 5.  (4)에서 추출된 데이터셋에서 나이그룹 속성으로 그룹을 분리

```python
grouped=obese.groupby('ageGroup')
```

### 6. (5)그룹에 대한 요약 통계

```python
grouped.describe()
```

### 7. 남녀 각각 20명씩 추출하여 하나의 데이터 프레임으로 결합 , 요약통계확인

```python
dt=data.sample(n=20)
display(dt)
dt.describe()
```
